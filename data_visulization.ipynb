# Import pandas matplotlib seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# load the dataset

```python
df = pd.read_csv('Practice-Datasets .csv')
```

# 1. Create a histogram to analyze the distribution of Age

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=15, kde=True, color='maroon', edgecolor='black', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# 2. Plot a pie chart showing the percentage distribution of Gender

```python
plt.figure(figsize=(6, 6))
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'pink'], explode=(0.05, 0))
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

# 3. Create a bar chart showing the number of customers in each City

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='City', data=df, hue='City', palette='viridis', legend=False, edgecolor='black', order=df['City'].value_counts().index)
plt.title('Number of Customers per City')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

# 4. Plot a box plot to analyze Monthly Income distribution

```python
plt.figure(figsize=(5, 7))
# Plot a box plot to analyze Monthly Income distribution
sns.boxplot(y=df['Monthly_Income'], color='cyan', width=0.4, linewidth=2)
plt.title('Monthly Income Distribution')
plt.ylabel('Monthly Income (INR)')
plt.show()
```

# 5. Create a histogram of Spending Score and interpret the skewness

```python
plt.figure(figsize=(8, 5))
# Create a histogram of Spending Score and interpret the skewness
sns.histplot(df['Spending_Score'], bins=20, color='purple', edgecolor='white', kde=True)
plt.title('Spending Score Distribution')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Count')
plt.show()
```

# 6. Plot a line chart to visualize Purchase Amount trend over Purchase Date

```python
plt.figure(figsize=(12, 5))
# Plot a line chart to visualize Purchase Amount trend over Purchase Date
df.groupby('Purchase_Date')['Purchase_Amount'].sum().plot(color='teal', linewidth=2, marker='o', markersize=4)
plt.title('Total Purchase Amount Trend')
plt.xlabel('Purchase Date')
plt.ylabel('Total Purchase Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df_trend = df.set_index('Purchase_Date')['Purchase_Amount'].resample('M').sum()
plt.plot(df_trend.index, df_trend.values, color='teal', linewidth=2, marker='o', markerfacecolor='maroon', markersize=6)
plt.title('Monthly Purchase Amount Trend')
plt.xlabel('Date (Monthly)')
plt.ylabel('Total Purchase Amount')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# 7. Create a bar plot to compare average Monthly Income by Gender

```python
plt.figure(figsize=(7, 5))
sns.barplot(x='Gender', y='Monthly_Income', data=df)
plt.title('Average Monthly Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Income')
plt.show()
```

# 8. Plot a box plot of Purchase Amount grouped by Product Category

```python
plt.figure(figsize=(12, 6))
# Corrected: Assigned 'Product_Category' to hue and set legend=False
sns.boxplot(x='Product_Category', y='Purchase_Amount', data=df)
plt.title('Purchase Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Purchase Amount')
plt.show()
```

# 9. Create a pie chart to show Product Category distribution

```python
plt.figure(figsize=(7, 7))
df['Product_Category'].value_counts().plot.pie(autopct='%1.1f%%', cmap='tab10', startangle=140)
plt.title('Product Category Distribution')
plt.ylabel('')
plt.show()
```

# 10. Plot a histogram of Purchase Amount with appropriate bin size 

```python
plt.figure(figsize=(9, 5))
sns.histplot(df['Purchase_Amount'], bins=30, color='orange', edgecolor='black', alpha=0.6)
plt.title('Purchase Amount Frequency')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

# 11. Create a bar chart showing average Spending Score by City

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='Spending_Score', data=df)
plt.title('Average Spending Score per City')
plt.xlabel('City')
plt.ylabel('Avg Spending Score')
plt.show()
```

 # 12. Plot a line chart showing total Purchase Amount per month

```python
plt.figure(figsize=(10, 5))
df.set_index('Purchase_Date').resample('M')['Purchase_Amount'].sum().plot(kind='line', marker='s', color='darkgreen', linestyle='--')
plt.title('Total Monthly Purchase Amount')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount')
plt.show()
```

# 13. Create a box plot comparing Monthly Income across different Cities

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='City', y='Monthly_Income', data=df)
plt.title('Monthly Income Distribution by City')
plt.xlabel('City')
plt.ylabel('Monthly Income')
plt.show()
```

# 14. Plot a histogram to analyze Age distribution by Gender

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Age', hue='Gender', multiple='dodge', bins=10, palette='Set1', edgecolor='black')
plt.title('Age Distribution Segmented by Gender')
plt.xlabel('Age')
plt.ylabel('Customer Count')
plt.show()
```

# 15. Create a bar chart showing total Purchase Amount by Product Category

```python
plt.figure(figsize=(10, 6))
total_sales = df.groupby('Product_Category')['Purchase_Amount'].sum().reset_index()
sns.barplot(x='Product_Category', y='Purchase_Amount', data=total_sales)
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.show()
```

# 16. Plot a scatter plot between Monthly Income and Spending Score

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x='Monthly_Income', y='Spending_Score', data=df, color='maroon', edgecolor='black',  marker='o')
plt.title('Monthly Income vs Spending Score')
plt.xlabel('Monthly Income')
plt.ylabel('Spending Score')
plt.show()
```

# 17. Create a line chart to visualize daily Purchase Amount variation

```python
plt.figure(figsize=(14, 6))
df.groupby('Purchase_Date')['Purchase_Amount'].sum().plot(color='black')
plt.title('Daily Variation in Purchase Amount')
plt.xlabel('Date')
plt.ylabel('Purchase Amount')
plt.show()
```

# 18. Plot a box plot of Spending Score grouped by Gender

```python
plt.figure(figsize=(7, 6))
sns.boxplot(x='Gender', y='Spending_Score', data=df, hue='Gender', palette='husl', width=0.5, legend=False)
plt.title('Spending Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Spending Score')
plt.show()
```

# 19. Create a bar chart showing customer count by Product Category and Gender

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Product_Category', hue='Gender', data=df, palette='viridis', edgecolor='black')
plt.title('Customer Count per Category Segmented by Gender')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

# 20. Analyze the relationship between Age and Purchase Amount using a suitable visualization

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Purchase_Amount', data=df, scatter_kws={'color':'darkblue', 'alpha':0.3}, line_kws={'color':'red'})
plt.title('Relationship: Age vs Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.show()
```
